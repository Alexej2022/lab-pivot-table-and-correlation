#### 4 We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.

```python
employee =  pd.read_csv("Employee.csv")
employee.head(11)
```

```python
# 4 We want to display the 
#number of years of work experience and the 
#salary for each Title. 
#Add Years as a column in your pivot table.

df4 = employee.pivot_table(index= ['Department','Title'], values = ['Salary','Years'], aggfunc = {'Salary':np.mean,'Years':np.sum}).reset_index()

df4.head(10)
```
